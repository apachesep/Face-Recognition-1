```python
from PIL import Image, ImageDraw
import numpy as np
from matplotlib import pyplot as plt
import cv2
import face_recognition
```

# 使用不同演算法去提取特徵

* detector: 檢測器，返回Keypoints
* descriptor: 從Keypoints結果得到描述子Mat對象

## 特徵提取方法
- BRISK: detector + descriptor
- ORB: detector + descriptor
- MSER: detector
- FAST: detector
- AGAST: detector
- GFFT: detector
- SimpleBlobDetector: detector
- KAZE: detector + descriptor
- AKAZE: detector + descriptor
- FREAK: descriptor
- StarDetector: detector
- BriefDescriptorExtractor: descriptor
- LUCID: descriptor
- LATCH: descriptor
- DAISY: descriptor
- MSDDetector: detector
- SIFT: detector + descriptor
- SURF: detector + descriptor

```python
# SIFT

IMG_URL = 'images/eye_1.jpg'
img = cv2.imread(IMG_URL)
gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

dummy = np.zeros((1,1))
sift = cv2.xfeatures2d.SIFT_create()
keypoints1, descriptors1 = sift.detectAndCompute(gray_image, None)
print('SIFT features',len(keypoints1))
img = cv2.drawKeypoints(img, keypoints1, dummy, flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
# cv2.imshow('SIFT features', img)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.show()
```

```python
# SURF

IMG_URL = 'images/eye_1.jpg'
img = cv2.imread(IMG_URL)
gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

dummy = np.zeros((1,1))
surf = cv2.xfeatures2d.SURF_create()
keypoints1, descriptors1 = surf.detectAndCompute(gray_image, None)
print('SURF features',len(keypoints1))

img = cv2.drawKeypoints(img, keypoints1, dummy, flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)

img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.show()
```

```python
# MSER not have descriptor use BRIEF

IMG_URL = 'images/eye_1.jpg'
img = cv2.imread(IMG_URL)
gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

dummy = np.zeros((1,1))
mser = cv2.MSER_create()

# Initiate BRIEF extractor
brief = cv2.xfeatures2d.BriefDescriptorExtractor_create()
keypoints = mser.detect(gray_image)
print('MSER_create:',len(keypoints))

# compute the descriptors with BRIEF
keypoints, descriptors = brief.compute(gray_image, keypoints)
img = cv2.drawKeypoints(img, keypoints, dummy, flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)

img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.show()
```

```python
# MSER

IMG_URL = 'images/eye_1.jpg'
img = cv2.imread(IMG_URL)
gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

mser = cv2.MSER_create()
keypoints = mser.detect(gray_image)
print('MSER_create:',len(keypoints))

img = cv2.drawKeypoints(img, keypoints, dummy, flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)

img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.show()
```
